```python
import os
import matplotlib.pyplot as plt
import numpy as np
import random

import torch

import gym
import ray

from ray.tune.logger import pretty_print
from ray import tune

from ray.rllib.agents import ppo, sac, ddpg

ray.init(ignore_reinit_error=True, log_to_driver=False)
```

# Create Continous Gym Environment




```python
# LunarLanderContinuous-v2
# MountainCarContinuous-v0
# Pendulum-v0

env_name = "LunarLanderContinuous-v2"
#env = gym.make(env_name)

#log_dir = f"./gym/{env_name}"
#os.makedirs(log_dir, exist_ok=True)
```

```python
config = sac.DEFAULT_CONFIG.copy()
config["use_pytorch"] = True
config["train_batch_size"] = tune.sample_from(lambda spec: random.choice([100, 256, 1000]))
config["num_gpus"]=1
config["env"]=env_name
```

```python
agent = sac.SACTrainer(config=config)
```

```python
sac_analysis = tune.run("SAC",config=config)

# for i in range(10):
#     result = agent.train()
#     print(pretty_print(result))
```
